```python
# imports

import matplotlib.pyplot as plt
from sklearn.neural_network import MLPClassifier
from sklearn import model_selection, metrics
from imblearn.over_sampling import SMOTE
import csv
import random
```

```python
# general specifications

path = "data/"
model_1 = "single_model/snap_model"
model_2 = "combined_model/mm3"
sw = ""
model_1_path = path + model_1 + sw
model_2_path = path + model_2 + sw

plot_path = "D:/Dropbox/masterthesis/thesis/plots/machine_learning/nn_analysis/"
```

```python
# import data

positives_dna_1 = []
negatives_dna_1 = []
positives_enzyme_1 = []
negatives_enzyme_1 = []

with open(model_1_path+'_pdidb_positives.txt') as p_dna:
    for row in csv.reader(p_dna,delimiter="\t"):
        positives_dna_1.append(row)

with open(model_1_path+'_pdidb_negatives.txt') as n_dna:
    for row in csv.reader(n_dna,delimiter="\t"):
        negatives_dna_1.append(row)
       
with open(model_1_path+'_enzyme_positives.txt') as p_enzyme:
    for row in csv.reader(p_enzyme,delimiter="\t"):
        positives_enzyme_1.append(row)
        
with open(model_1_path+'_enzyme_negatives.txt') as n_enzyme:
    for row in csv.reader(n_enzyme,delimiter="\t"):
        negatives_enzyme_1.append(row)
        
positives_dna_2 = []
negatives_dna_2 = []
positives_enzyme_2 = []
negatives_enzyme_2 = []

with open(model_2_path+'_pdidb_positives.txt') as p_dna:
    for row in csv.reader(p_dna,delimiter="\t"):
        positives_dna_2.append(row)

with open(model_2_path+'_pdidb_negatives.txt') as n_dna:
    for row in csv.reader(n_dna,delimiter="\t"):
        negatives_dna_2.append(row)
       
with open(model_2_path+'_enzyme_positives.txt') as p_enzyme:
    for row in csv.reader(p_enzyme,delimiter="\t"):
        positives_enzyme_2.append(row)
        
with open(model_2_path+'_enzyme_negatives.txt') as n_enzyme:
    for row in csv.reader(n_enzyme,delimiter="\t"):
        negatives_enzyme_2.append(row)
```

```python
# pre-process data

tmp_dna_1 = positives_dna_1 + negatives_dna_1
tmp_enzyme_1 = positives_enzyme_1 + negatives_enzyme_1

labels_dna_1 = [1]*len(positives_dna_1)+[0]*len(negatives_dna_1)
labels_enzyme_1 = [1]*len(positives_enzyme_1)+[0]*len(negatives_enzyme_1)

dna_1 = []
enzyme_1 = []

for j in tmp_dna_1:
    new_j = [float(i) for i in j]
    dna_1.append(new_j)
    
for j in tmp_enzyme_1:
    new_j = [float(i) for i in j]
    enzyme_1.append(new_j)

tmp_dna_2 = positives_dna_2 + negatives_dna_2
tmp_enzyme_2 = positives_enzyme_2 + negatives_enzyme_2

labels_dna_2 = [1]*len(positives_dna_2)+[0]*len(negatives_dna_2)
labels_enzyme_2 = [1]*len(positives_enzyme_2)+[0]*len(negatives_enzyme_2)

dna_2 = []
enzyme_2 = []

for j in tmp_dna_2:
    new_j = [float(i) for i in j]
    dna_2.append(new_j)
    
for j in tmp_enzyme_2:
    new_j = [float(i) for i in j]
    enzyme_2.append(new_j)
```

```python
dna_x_train_1,dna_x_test_1,dna_y_train_1,dna_y_test_1 = model_selection.train_test_split(dna_1, labels_dna_1, test_size=0.2, random_state=6)
enzyme_x_train_1,enzyme_x_test_1,enzyme_y_train_1,enzyme_y_test_1 = model_selection.train_test_split(enzyme_1,labels_enzyme_1,test_size=0.2,random_state=6)

tmp_x_train_1 = dna_x_train_1 + enzyme_x_train_1
tmp_y_train_1 = dna_y_train_1 + enzyme_y_train_1

x_train_1,x_cross_1,y_train_1,y_cross_1 = model_selection.train_test_split(tmp_x_train_1,tmp_y_train_1,test_size=0.2,random_state=6)

sm = SMOTE(random_state=42)
x_train_1, y_train_1 = sm.fit_sample(x_train_1,y_train_1)

dna_x_train_2,dna_x_test_2,dna_y_train_2,dna_y_test_2 = model_selection.train_test_split(dna_2, labels_dna_2, test_size=0.2, random_state=6)
enzyme_x_train_2,enzyme_x_test_2,enzyme_y_train_2,enzyme_y_test_2 = model_selection.train_test_split(enzyme_2,labels_enzyme_2,test_size=0.2,random_state=6)

tmp_x_train_2 = dna_x_train_2 + enzyme_x_train_2
tmp_y_train_2 = dna_y_train_2 + enzyme_y_train_2

x_train_2,x_cross_2,y_train_2,y_cross_2 = model_selection.train_test_split(tmp_x_train_2,tmp_y_train_2,test_size=0.2,random_state=6)

sm = SMOTE(random_state=42)
x_train_2, y_train_2 = sm.fit_sample(x_train_2,y_train_2)
```

```python
# generate random labels

y_train_rnd = random.shuffle(y_train_1)
y_cross_rnd = random.shuffle(y_cross_1)
```

```python
# calculate training and validation scores

iterations = (20, 40, 60, 80, 100, 120, 140, 160, 300)

train_scores_1 = []
test_scores_1 = []
train_scores_2 = []
test_scores_2 = []
train_scores_rnd = []
test_scores_rnd = []
    
for iter in iterations:
    print(iter)
    classifier1 = MLPClassifier(hidden_layer_sizes=(200,),max_iter=iter,tol=-100)   
    classifier1.fit(x_train_1,y_train_1)
        
    train_score_1 = classifier1.score(x_train_1,y_train_1)
    test_score_1 = classifier1.score(x_cross_1,y_cross_1)
        
    train_scores_1.append(train_score_1)
    test_scores_1.append(test_score_1)
    
    classifier2 = MLPClassifier(hidden_layer_sizes=(200,),max_iter=iter,tol=-100)   
    classifier2.fit(x_train_2,y_train_2)
        
    train_score_2 = classifier1.score(x_train_2,y_train_2)
    test_score_2 = classifier1.score(x_cross_2,y_cross_2)
        
    train_scores_2.append(train_score_2)
    test_scores_2.append(test_score_2)
    
    classifier1 = MLPClassifier(hidden_layer_sizes=(200,),max_iter=iter,tol=-100)   
    classifier1.fit(x_train_1,y_train_1)
        
    train_score_rnd = classifier1.score(x_train_1,y_train_rnd)
    test_score_rnd = classifier1.score(x_cross_1,y_cross_rnd)
        
    train_scores_rnd.append(train_score_rnd)
    test_scores_rnd.append(test_score_rnd)


# plot figure

fig = plt.figure()

    plt.title("Validation Curves")
    plt.xlabel("Iterations")
    plt.ylabel("Accuracy")
    plt.ylim(0.0, 1.0)
    lw = 2
    plt.plot(iterations, train_scores_1, label="Training score MM3",
             color="#ff0000", lw=lw)
    plt.plot(iterations, test_scores_1, label="Cross-validation score MM3",
             color="#ff6666", lw=lw)
    plt.plot(iterations, train_scores_2, label="Training score SNAP",
             color="#00ff00", lw=lw)
    plt.plot(iterations, test_scores_2, label="Cross-validation score SNAP",
             color="#66ff66", lw=lw)
    plt.plot(iterations, train_scores_rnd, label="Training score RANDOM",
             color="#0000ff", lw=lw)
    plt.plot(iterations, test_scores_rnd, label="Cross-validation score RANDOM",
             color="#6666ff", lw=lw)
    plt.legend(loc="best")

    fig.savefig(plot_path+"model_comparison.png")
    plt.close("all")
```
